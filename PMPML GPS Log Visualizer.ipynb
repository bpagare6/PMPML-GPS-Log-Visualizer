<h3>Problem Statement: PMPML GPS Log Visualizer</h3>
<h4>Approach:</h4>
<ol>
    <li>Make the clear Data Set by removing the irrelevent part</li>
    <li>Filter the Data Set with respect to One PMPML route-id</li>
    <li>Then Sorted the New Data Set as per Time Frame</li>
    <li>Plotted each point on the Google Map</li>
    <li>Then Created the Animation from it</li>
</ol>

```python
# Imported tarfile for Extracting files in the tar file
import tarfile

# Opened the tar.gz file
tf = tarfile.open('PMPML_GPSLogs.tar.gz')

# Files extracted
tf.extractall()
```

```python
# Imported Pandas for Reading CSV files
import pandas as pd

# df is DataFrame
df = pd.read_csv('PMPML_GPSLogs.csv', dtype = str)
```

```python
# Shows sample 5 entries of the DataFrame
df.sample(5)
```

```python
# Creating Useful data by droping Unimportant columns
unimportant_cols = ['sr', 'feed_timestamp', 'id','stop_id',\
                    'file', 'file_sr', 'scheduleRelationship',\
                    'year_month', 'date', 'hour',\
                    'minute', 'second', 'year', 'month',\
                    'currentStopSequence', 'currentStatus']

df.drop(unimportant_cols, axis = 1, inplace = True)
```

```python
# Show sample 5 entries of modified DataFrame
df.sample(5)
```

```python
# Selecting a Single Bus route and Single Vehicle ID
df = df.loc[df['route_id'] == '42'].loc[df['vehicle_id'] == '1661']
```

```python
# Sample 5 entries of modified DataFrame
df.sample(5)
```

```python
# Saved the new DataFrame created for Refining
df.to_csv('PMPML_GPSLogs_For_route42_vehicleid1661_onebus.csv', sep = ',')
```

<h6>This file PMPML_GPSLogs_For_route42_vehicleid1661_onebus.csv was Refined by using Google OpenRefine</h6>
<h6>And the Refined file is PMPML_GPSLogs_For_route42_vehicleid1661_onebus_refined.csv</h6>

```python
# Loading new DataSet
df = pd.read_csv('PMPML_GPSLogs_For_route42_vehicleid1661_onebus_refined.csv', dtype = str)

# Sample after refining
df.sample()
```

```python
# The DataFrame is sorted with various Parameters
df.sort_values(by = ['From', 'Date', 'Time', 'Time of Leave'], inplace = True)
```

```python
# The new DataFrame looks like,
df
```

```python
# Now Selecting only one Day and One Single Bus
df = df.loc[df['Date'] == '2019-01-19'].loc[df['Time of Leave'] == '15:45']

# Creating Final file for Reference
df.to_csv('PMPML_GPSLogs_For_route42_vehicleid1661_onebus_refined_final.csv', sep = ',')

# Final DataFrame Looks like,
df
```

```python
# Importing Folium for Map Visualization
import folium

# Creating new map with respect to Longitude and Latitude
GPSLogs = folium.Map(location = [float(df["latitude"][0]), float(df["longitude"][0])])
```

```python
# Add Markers in the Map
zoom_start = 13

tiles = 'PMPML GPS Logs Route:42'

for i in range(162):
    marker = folium.CircleMarker(location = [float(df["latitude"][i]), float(df["longitude"][i])])
    marker.add_to(GPSLogs)
```

```python
# Finally Saving the Map in the html file
GPSLogs.save('GPSLogs.html')
```
